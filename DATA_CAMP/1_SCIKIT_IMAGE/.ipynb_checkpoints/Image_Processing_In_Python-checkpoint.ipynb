```python
import skimage
import numpy             as np
import pandas            as pd
import matplotlib.pyplot as plt
```

```python
## importing defult image
coffee_image = skimage.data.coffee()
coin_image   = skimage.data.coins()
```

```python
def image_description(image):
    print("--" * 20)
    print("SHAPE : {}".format(image.shape))
    print("TYPE  : {}".format(type(image)))
    print("SIZE  : {}".format(image.size))
    print("--" * 20)
def plot_image(image,title="Image"):
    plt.title(title)
    plt.imshow(image)
    image_description(image)
    
```

```python
plot_image(coffee_image)
```

```python
plot_image(coin_image)
```

```python
from skimage import data,color
rocket = data.rocket()
gray_scaled_rocket = color.rgb2gray(rocket)
plot_image(rocket)
```

```python
plot_image(gray_scaled_rocket)
```

```python
## flipping the image upside down
flipped_image_ud = np.flipud(gray_scaled_rocket)
plot_image(flipped_image_ud)
```

```python
## flipping the image horizontallly
flipped_image_lr = np.fliplr(gray_scaled_rocket)
plot_image(flipped_image_lr)
```

```python
red_channel = coffee_image[:,:,0]
plot_image(red_channel)
```

## Image Thresholding

```python
## to use the thresh holding you must conver t the image into threshholding
gray = data.camera()
thresh = 127
#gray = skimage.color.rgb2gray(image)
binary = gray > thresh
plot_image(binary)
```

```python
plot_image(gray)
```

```python
## inverted thresholding
gray = data.camera()
thresh = 127
#gray = skimage.color.rgb2gray(image)
binary = gray <= thresh
plot_image(binary)
```
