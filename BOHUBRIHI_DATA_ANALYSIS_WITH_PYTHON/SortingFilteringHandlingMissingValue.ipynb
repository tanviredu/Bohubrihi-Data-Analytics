## .where() and .filter()

```python
import numpy as np
```

```python
df = pd.DataFrame(np.random.randint(1,100,[10,5]),columns=["id","Test_1","Test_2","Test_3","Test_5"])
```

```python
df
```

```python
df['Test_3']
```

```python
df['Test_3'].where(cond=df['Test_3']>=20,other=0)
```

```python
df.filter(items=['Test_1',"Test_2"])
```

```python
df.filter(like="Test")
```

```python
df.filter(regex="Test")
```

```python
df.filter(regex="[0-3]")
```
