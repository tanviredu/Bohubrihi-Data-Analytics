```python
## Join Query in Pandas
## suppose we need to make join query in two table
## first make two table
```

```python
d1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
```

```python
d1
```

```python
d2 = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2']})
```

```python
d2
```

```python
## we want to join d1 and d2 in table based in the key column
## we need table nameand key based on which it will join
## THIS IS LEFT JOIN
joined_table = d1.join(d2.set_index('key'),on="key") 
```

```python
joined_table
```

```python
## MAKE A INNER JOIN
INNER_JOIN_TABLE = pd.merge(d1,d2,on="key",how="inner")
```

```python
INNER_JOIN_TABLE = INNER_JOIN_TABLE.set_index('key')
```

```python
INNER_JOIN_TABLE
```

```python
## make a left join with merge
LEFT_JOIN_TABLE = pd.merge(d1,d2,on="key",how="left")
LEFT_JOIN_TABLE = LEFT_JOIN_TABLE.set_index("key")
LEFT_JOIN_TABLE
```

```python
## make a right join with merge
RIGHT_JOIN_TABLE = pd.merge(d1,d2,on="key",how="right")
RIGHT_JOIN_TABLE = RIGHT_JOIN_TABLE.set_index("key")
RIGHT_JOIN_TABLE
```

```python
## make a OUTER join with merge
OUTER_JOIN_TABLE = pd.merge(d1,d2,on="key",how="outer")
OUTER_JOIN_TABLE = OUTER_JOIN_TABLE.set_index("key")
OUTER_JOIN_TABLE
```
