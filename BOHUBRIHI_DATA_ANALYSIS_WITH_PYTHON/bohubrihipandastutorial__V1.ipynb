```python
## Join Query in Pandas
## suppose we need to make join query in two table
## first make two table
```

```python
d1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
```

```python
d1
```

```python
d2 = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2']})
```

```python
d2
```

```python
## we want to join d1 and d2 in table based in the key column
## we need table nameand key based on which it will join
## THIS IS LEFT JOIN
joined_table = d1.join(d2.set_index('key'),on="key") 
```

```python
joined_table
```

```python
## MAKE A INNER JOIN
INNER_JOIN_TABLE = pd.merge(d1,d2,on="key",how="inner")
```

```python
INNER_JOIN_TABLE = INNER_JOIN_TABLE.set_index('key')
```

```python
INNER_JOIN_TABLE
```

```python
## make a left join with merge
LEFT_JOIN_TABLE = pd.merge(d1,d2,on="key",how="left")
LEFT_JOIN_TABLE = LEFT_JOIN_TABLE.set_index("key")
LEFT_JOIN_TABLE
```

```python
## make a right join with merge
RIGHT_JOIN_TABLE = pd.merge(d1,d2,on="key",how="right")
RIGHT_JOIN_TABLE = RIGHT_JOIN_TABLE.set_index("key")
RIGHT_JOIN_TABLE
```

```python
## make a OUTER join with merge
OUTER_JOIN_TABLE = pd.merge(d1,d2,on="key",how="outer")
OUTER_JOIN_TABLE = OUTER_JOIN_TABLE.set_index("key")
OUTER_JOIN_TABLE
```

```python
## Suppose we want the row 1, 3,5  and their grade and Test-Score
## The code is
df.loc[[1,3,5],['Grade','Test-Score']]
```

```python
## but if we want all the column of the [1,2,5] 
##  we need to use the : wild card . it means all
df.loc[[1,3,5],:]
```

```python
## Suppose we need Name and City column and 10 of the row
## from 1 to 10, we can do with the  this code
df.loc[1:10,['Name','City']]
```

```python
## Suppose we need Name and City column and the row whose Row Id is divisible by 3
## we can do with the  this code
df.loc[[i for i in df.index if i%3==0],['Name','City']]

## [i for i in df.index if i%3==0] 
## this is called list comprehension
```
