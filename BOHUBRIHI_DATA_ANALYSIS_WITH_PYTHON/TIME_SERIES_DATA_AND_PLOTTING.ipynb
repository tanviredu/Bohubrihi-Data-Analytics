```python
df_test = pd.read_csv("./DailyDelhiClimateTest.csv",index_col="date")
```

```python
### requirment give me the average tempareture of every 7 days
df_test.index = pd.to_datetime(df_test.index)
```

```python
## every week's average tempareture
## we resample the data every sample takes 7 days
df_test[['tmp']].resample(rule="W").mean()
```

```python
df_test[['tmp']].plot()  ## every day temperature plot
df_test[['tmp']].resample(rule="W").mean().plot() ## every week avg tempareture plot
df_test[['tmp']].resample(rule="M").mean().plot() ## every month avg temp plot
plt.figure()
plt.show()
```

```python
## find every weeks average humdity from the dataset
## remember resample() function is like the group by aggragate method
## it combines 7 days (for week) and then find the mean of the data within 7 days
df_test[['humidity']].plot()
df_test[['humidity']].resample(rule="W").mean().plot()
```

```python
df_test[['wind_speed']].plot()
df_test[['wind_speed']].resample(rule="W").mean().plot()
df_test[['wind_speed']].resample(rule="M").mean().plot()
```
