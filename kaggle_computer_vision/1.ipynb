```python
import os, warnings
import matplotlib.pyplot as plt
import numpy             as np
import tensorflow        as tf
from matplotlib                     import gridspec
from tensorflow.keras.preprocessing import image_dataset_from_directory
warnings.filterwarnings("ignore") # to clean up output cells
```

```python
# Reproducability
def set_seed(seed=31415):
    np.random.seed(seed)
    tf.random.set_seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    os.environ['TF_DETERMINISTIC_OPS'] = '1'
set_seed()

# Set Matplotlib defaults
plt.rc('figure', autolayout=True)
plt.rc('axes', labelweight='bold', labelsize='large',
       titleweight='bold', titlesize=18, titlepad=10)
plt.rc('image', cmap='magma')
```

ds_train_: This is presumably a variable that will hold the training dataset created by the image_dataset_from_directory function.

image_dataset_from_directory: This is a function that is commonly used in machine learning libraries like TensorFlow and Keras for loading image datasets from a directory. It simplifies the process of creating datasets for image classification tasks.

'../input/car-or-truck/train': This is the path to the directory containing the training images. The code expects the images to be organized into subdirectories, where each subdirectory corresponds to a different class. In this case, it seems to be a binary classification problem with classes "car" and "truck."

labels='inferred': This parameter tells the function to automatically infer the class labels from the subdirectories' names. In this case, it will likely set 'car' to label 0 and 'truck' to label 1.

label_mode='binary': This parameter specifies that the labels are binary, indicating the two classes, 'car' and 'truck.'

image_size=[128, 128]: This parameter sets the size to which the images will be resized. All images in the dataset will be resized to a resolution of 128x128 pixels.

interpolation='nearest': This parameter specifies the method used for resizing the images. In this case, 'nearest' likely refers to nearest-neighbor interpolation, which is a simple but fast method for resizing images.

batch_size=64: This parameter determines how many images will be loaded in each batch during training. It sets the batch size to 64, meaning that the model will be updated with gradients computed from 64 images at a time.

shuffle=True: This parameter indicates that the dataset should be shuffled. Shuffling the dataset is important during training to introduce randomness and prevent the model from learning patterns related to the order of the images.

```python
# Load training and validation sets
ds_train_ = image_dataset_from_directory(
    '../input/car-or-truck/train',
    labels='inferred',
    label_mode='binary',
    image_size=[128, 128],
    interpolation='nearest',
    batch_size=64,
    shuffle=True,
)
ds_valid_ = image_dataset_from_directory(
    '../input/car-or-truck/valid',
    labels='inferred',
    label_mode='binary',
    image_size=[128, 128],
    interpolation='nearest',
    batch_size=64,
    shuffle=False,
)
```

```python
# Converting method to convert pixel in float
def convert_to_float(image, label):
    image = tf.image.convert_image_dtype(image, dtype=tf.float32)
    return image, label
```

In the convert_to_float function, the arguments image and label represent the elements of each data point in the dataset. When you use .map(convert_to_float), TensorFlow will call convert_to_float with these elements, and the function will process them accordingly.

Regarding the labels in the ds_train_ dataset, you can typically assume that labels are part of the dataset if you've used the image_dataset_from_directory function with the labels='inferred' parameter. This means that the labels are inferred from the directory structure where subdirectories represent different classes, and the names of the subdirectories are used as labels. In your original code, it was set up like this:

```python
## Data PipeLine
ds_train = (
    ds_train_
    .map(convert_to_float)
    .cache()
)
ds_valid = (
    ds_valid_
    .map(convert_to_float)
    .cache()
)
```

The **InceptionV1** model pretrained on ImageNet is available in the [TensorFlow Hub](https://www.tensorflow.org/hub/) repository, but we'll load it from a local copy. Run this cell to load InceptionV1 for your base.

```python
import tensorflow_hub as hub

pretrained_base = tf.keras.models.load_model(
    '../input/cv-course-models/cv-course-models/inceptionv1'
)
```

Now that the base is defined to do the feature extraction, create a head of `Dense` layers to perform the classification, following this diagram:

```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    pretrained_base,
    layers.Flatten(),
    layers.Dense(6,activation="relu"),
    layers.Dense(1,activation="sigmoid")
])

# Check your answer
q_2.check()
```

# 3) Train #

Before training a model in Keras, you need to specify an *optimizer* to perform the gradient descent, a *loss function* to be minimized, and (optionally) any *performance metrics*. The optimization algorithm we'll use for this course is called ["Adam"](https://keras.io/api/optimizers/adam/), which generally performs well regardless of what kind of problem you're trying to solve.

The loss and the metrics, however, need to match the kind of problem you're trying to solve. Our problem is a **binary classification** problem: `Car` coded as 0, and `Truck` coded as 1. Choose an appropriate loss and an appropriate accuracy metric for binary classification.

```python
# YOUR CODE HERE: what loss function should you use for a binary
# classification problem? (Your answer for each should be a string.)
optimizer = tf.keras.optimizers.Adam(epsilon=0.01)
model.compile(
    optimizer=optimizer,
    loss = "binary_crossentropy",
    metrics=["binary_accuracy"],
)

# Check your answer
q_3.check()
```

```python
# Lines below will give you a hint or solution code
#q_3.hint()
#q_3.solution()
```

```python
history = model.fit(
    ds_train,
    validation_data=ds_valid,
    epochs=30,
)
```

Run the cell below to plot the loss and metric curves for this training run.

```python
import pandas as pd
history_frame = pd.DataFrame(history.history)
history_frame.loc[:, ['loss', 'val_loss']].plot()
history_frame.loc[:, ['binary_accuracy', 'val_binary_accuracy']].plot();
```
